### 1.1 Definición clara del problema de investigación

El problema de investigación se centra en el desarrollo de un modelo de clasificación supervisada capaz de determinar si un individuo padece o no una enfermedad cardíaca, a partir de un conjunto de variables clínicas y fisiológicas, tales como el nivel de colesterol, glucosa, el género, la altura, el peso y los hábitos asociados a la salud del paciente. En términos formales, se busca modelar la relación existente entre un vector de características y una variable objetivo binaria, donde cada valor representa la ausencia o presencia de enfermedad cardíaca, respectivamente.

Este problema resulta especialmente relevante debido a la alta prevalencia de las enfermedades cardiovasculares a nivel global, las cuales constituyen una de las principales causas de mortalidad. En este contexto, la capacidad de identificar de manera temprana a individuos con mayor probabilidad de desarrollar estas patologías representa un avance significativo en términos de prevención, diagnóstico oportuno y optimización de recursos en los sistemas de salud. Un modelo predictivo confiable puede servir como herramienta de apoyo en la toma de decisiones clínicas, permitiendo priorizar la atención de pacientes en riesgo y reducir la carga sobre los profesionales de la salud.

La importancia de contar con un modelo que funcione correctamente radica en la naturaleza crítica del dominio de aplicación. Un desempeño deficiente podría derivar en falsos negativos, omitiendo diagnósticos en pacientes que requieren atención, o en falsos positivos, generando intervenciones innecesarias y costos adicionales. 

En este escenario, la inteligencia artificial y, en particular, las técnicas de aprendizaje automático, ofrecen un enfoque robusto para abordar este tipo de problemáticas. Estas herramientas permiten identificar patrones complejos y no lineales en los datos que pueden pasar desapercibidos mediante métodos tradicionales. Su implementación puede contribuir significativamente al fortalecimiento de los sistemas de salud, facilitando diagnósticos más rápidos y precisos, promoviendo la medicina preventiva y apoyando la toma de decisiones basada en datos. En consecuencia, el uso de modelos de inteligencia artificial en la predicción de enfermedades cardíacas no solo tiene un impacto técnico, sino también un alto potencial social al mejorar la calidad de vida de la población y reducir la mortalidad asociada a estas enfermedades.

#### 1.6.2 Outliers

La detección y tratamiento de valores atípicos (outliers) se llevó a cabo mediante el uso del método del rango intercuartílico, una técnica robusta ampliamente utilizada en análisis exploratorio de datos debido a que no requiere supuestos de normalidad en la distribución de las variables. Para cada una de las variables seleccionadas  se calcularon el primer cuartil y el tercer cuartil, considerando como outliers aquellas observaciones que se encontraban fuera de dicho intervalo.

En lugar de eliminar los registros que contenían valores atípicos, se optó por aplicar una estrategia de tratamiento basada en winsorización, específicamente mediante la técnica de clipping. Este procedimiento consiste en truncar los valores extremos, de modo que cualquier observación por debajo del límite inferior es reemplazada por dicho límite, y cualquier valor por encima del límite superior es ajustado al límite superior correspondiente. Esta decisión permite conservar la totalidad de los datos, evitando la pérdida de información potencialmente relevante, al tiempo que reduce la influencia desproporcionada de valores extremos sobre el comportamiento de los modelos.

Adicionalmente, se realizó una evaluación cuantitativa del proceso, contabilizando el número de outliers antes y después del tratamiento para cada variable. Como resultado de la aplicación del clipping, todos los valores quedaron acotados dentro del rango definido, eliminando efectivamente la presencia de valores atípicos en el conjunto de datos transformado.

#### 2.1.6 IQR

```python
print("=== Análisis de Outliers con IQR ===\n")

for col in columnas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    pct = (len(outliers) / len(df)) * 100

    print(f"{col}:")
    print(f"  Q1={Q1:.2f} | Q3={Q3:.2f} | IQR={IQR:.2f}")
    print(f"  Límite inferior={lower:.2f} | Límite superior={upper:.2f}")
    print(f"  Outliers: {len(outliers):,} ({pct:.2f}%)\n")
```

```python
cols_outlier = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
df_clean = df_imp.copy()

print(f'{"Variable":<12} {"Outliers antes":>15} {"Límite inf":>12} {"Límite sup":>12} {"Outliers después":>18}')
print('-' * 73)

for col in cols_outlier:
    Q1  = df_clean[col].quantile(0.25)
    Q3  = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    antes = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    df_clean[col] = df_clean[col].clip(lower=lower, upper=upper)
    despues = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()

    print(f'{col:<12} {antes:>15,} {lower:>12.2f} {upper:>12.2f} {despues:>18,}')
```

#### 2.3.3 Codificación Categórica

```python
df_enc = df_clean.copy()

# gender: inicialmente es 1=Mujer y 2=Hombre → 0=Mujer, 1=Hombre
df_enc['gender'] = df_enc['gender'].map({1: 0, 2: 1})

# cholesterol y gluc son ordinales (1 < 2 < 3), se conservan tal cual
# smoke, alco, active ya son binarias 0/1

print('✅ Codificación aplicada:')
print(df_enc[['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']].head())
```

#### 2.3.4 Transformaciones

```python
from sklearn.preprocessing import StandardScaler

df_trans = df_enc.copy()
cols_transform = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']

# LOG
print(f'{"Variable":<12} {"Skew original":>15} {"Skew log1p":>12} {"¿Aplicar?":>12}')
print('-' * 55)

log_cols = []
for col in cols_transform:
    s_orig = df_trans[col].skew()
    s_log  = np.log1p(df_trans[col].clip(lower=0)).skew()
    apply  = abs(s_orig) > 0.5 and abs(s_log) < abs(s_orig)
    print(f'{col:<12} {s_orig:>15.4f} {s_log:>12.4f} {str(apply):>12}')
    if apply:
        df_trans[col + '_log'] = np.log1p(df_trans[col].clip(lower=0))
        log_cols.append(col)

print(f'\nColumnas con log1p aplicado: {log_cols}')

#Escalamiento
cols_scale   = [c + '_log' if c in log_cols else c for c in cols_transform]
feature_cols = cols_scale + ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']

scaler   = StandardScaler()
X_scaled = scaler.fit_transform(df_trans[feature_cols])
df_scaled = pd.DataFrame(X_scaled, columns=feature_cols, index=df_trans.index)
df_scaled['cardio'] = df_trans['cardio'].values

print('=== Estadísticos después de StandardScaler ===')
print(df_scaled[feature_cols].describe().round(3))
```

```python
#Selección de variables significativas
from scipy.stats import pointbiserialr

significant_cols = []

print(f"{'Variable':<25} {'Correlación':>12} {'p-valor':>12} {'Significativa':>14}")
print('-' * 67)

for col in feature_cols:
    corr_val, p_val = pointbiserialr(df_scaled[col], df_scaled['cardio'])
    sig = p_val < 0.05
    print(f"{col:<25} {corr_val:>12.4f} {p_val:>12.4e} {str(sig):>14}")
    if sig:
        significant_cols.append(col)

print(f'\nVariables seleccionadas ({len(significant_cols)}): {significant_cols}')
```

```python
#Eliminación de multicolinealidad usando el Variance Inflation Factor

def VIF_calculation(X):
    VIF = pd.DataFrame()
    VIF['variable'] = X.columns
    VIF['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    VIF = VIF.sort_values('VIF', ascending=False).reset_index(drop=True)
    return VIF

def delete_multicollinearity(df, target_name, VIF_threshold):
    X = df.drop(target_name, axis=1)
    VIF_mat = VIF_calculation(X)
    n_VIF = VIF_mat['VIF'][0]
    if n_VIF <= VIF_threshold:
        print('No hay multicolinealidad')
    else:
        while n_VIF > VIF_threshold:
            print(f'Eliminando: {VIF_mat["variable"][0]}  (VIF={n_VIF:.2f})')
            X = X.drop(VIF_mat['variable'][0], axis=1)
            VIF_mat = VIF_calculation(X)
            n_VIF = VIF_mat['VIF'][0]
    display(VIF_mat)
    return X

# Construimos df_for_vif solo con las variables significativas + target
df_for_vif = df_scaled[significant_cols].copy()
df_for_vif['cardio'] = df_scaled['cardio'].values

ml_X = delete_multicollinearity(df_for_vif, 'cardio', VIF_threshold=10)
```

```python
#Matriz de correlacion luego del VIF
corr = ml_X.corr().round(2)
cmap = sns.diverging_palette(220, 10, as_cmap=True)

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.set(font_scale=1.4)
f, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={'shrink': .5}, annot=True)
plt.title('Matriz de correlación — Variables post-VIF', fontsize=13)
plt.tight_layout()
plt.show()
```
